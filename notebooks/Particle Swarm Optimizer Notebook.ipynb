# Trajectory 1

```python
targetPoints = [261]
# targetPoints = []

# targetPoints.append(np.array(range(260, 262)))
# targetPoints.append(np.array(range(420, 470)))

# targetPoints = np.hstack(targetPoints)
```

```python
variablePoints = np.zeros( ( np.size( trajectory1, 0 ) ), dtype = bool )
variablePoints[targetPoints] = True
```

```python
start_time = time.time()

correctedTraj1 = TIL.TrajectoryImprovement(trajectory1, variablePoints, optAlgorithm='pso')

exec_time1 = time.time() - start_time
```

```python
print(f'Execution time of first trajectory: {exec_time1}')
```

```python
tmp = trajectory1[:, 1:3]
cTraj1 = correctedTraj1[:, 1:3]
```

```python
avg_lat = np.mean([row[0] for row in tmp])
avg_lon = np.mean([row[1] for row in tmp])
```

```python
m = folium.Map(
    location=[avg_lat, avg_lon],
    tiles="cartodbpositron",
    zoom_start=5
)

aline1 = folium.PolyLine(tmp, color='blue')
m.add_child(aline1)

aline2 = folium.PolyLine(cTraj1, color='red')
m.add_child(aline2)

folium.Marker(location=[correctedTraj1[261][1], correctedTraj1[261][2]], icon=folium.Icon(color='darkblue', icon_color='white', icon='male', angle=0, prefix='fa')).add_to(m)

m
```

```python
m.save('./Results/Particle swarm optimization/Run 10/trajectory1')
```

```python
TIL.aux_CalculateTrajectoryError(trajectory1, variablePoints)
```

```python
TIL.aux_CalculateTrajectoryError(correctedTraj1, variablePoints)
```
