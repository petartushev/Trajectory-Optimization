```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 15
```

```python
x_axis_labels = ['Hill climb','Simulated annealing','Tabu search','Local beam search', 'Genetic algorithm', 'PSO', 'Ant colony']
```

# Hill climb

```python
x_axis_labels = [f'Run {n}' for n in range(1, 11)]

before_run = np.full(shape=10, fill_value=0.378, dtype=np.float64)
after_run = np.full(shape=10, fill_value=0.0, dtype=np.float64)
x_axis = np.arange(len(x_axis_labels))

delta_run = before_run-after_run

exec_time = [0.611, 0.437, 0.234, 0.124, 1.066, 1.101, 0.032, 0.029, 0.031, 0.241]


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1 = plt.bar(x_axis -.2, delta_run, .2)
ax2 = plt.bar(x_axis + .2, exec_time, .2)
  

# ax1.set_ylabel('Trajectory error', fontsize=20)
# ax2.set_ylabel('Execution time in seconds', fontsize=20)

ax1.set_label('Error corrected')
ax2.set_label('Execution time in seconds')

# plt.title('Amount of error correction after algorithm exectuion for each algorithm. More is better.', fontsize=20)
plt.xticks(x_axis, x_axis_labels, fontsize=20, rotation=90)

xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(len(x_axis))]
xlabs=[i/2 for i in range(len(x_axis))]

for i, v in enumerate(delta_run):
    plt.text(xlocs[i] - 1.25, v + 0.015, str(round(v, ndigits=3)), fontsize=16, rotation=90)
    

for i, v in enumerate(exec_time):
    plt.text(xlocs[i] - .85, v + 0.015, str(round(v, ndigits=3)), fontsize=16, rotation=90)

plt.legend(fontsize=16)

plt.savefig('hill_climb - amount of correction - exectuion time.png')
```
