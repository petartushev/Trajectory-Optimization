```python
trajectory1 = np.load( '/home/petar/Fakultet/Semester 8/Biologically inspired computing/Proekt/SampleTraj/Trj_0.npy' )
trajectory1 = pd.DataFrame(trajectory1)
trajectory1.columns = ['timestamp', 'latitude', 'longitude', 'altitude']
trajectory1 = trajectory1.drop_duplicates(subset=['latitude', 'longitude'], keep='first')
trajectory1.reset_index(inplace=True)
trajectory1.drop('index', axis=1, inplace=True)
trajectory1.astype(float)
trajectory1['sum_of_coords'] = trajectory1['latitude'] + trajectory1['longitude']


trajectory2 = np.load( '/home/petar/Fakultet/Semester 8/Biologically inspired computing/Proekt/SampleTraj/Trj_1.npy' )
trajectory2 = pd.DataFrame(trajectory2)
trajectory2.columns = ['timestamp', 'latitude', 'longitude', 'altitude']
trajectory2 = trajectory2.drop_duplicates(subset=['latitude', 'longitude'], keep='first')
trajectory2.reset_index(inplace=True)
trajectory2.drop('index', axis=1, inplace=True)
trajectory2.astype(float)
trajectory2['sum_of_coords'] = trajectory2['latitude'] + trajectory2['longitude']


trajectory3 = np.load( '/home/petar/Fakultet/Semester 8/Biologically inspired computing/Proekt/SampleTraj/Trj_2.npy' )
trajectory3 = pd.DataFrame(trajectory3)
trajectory3.columns = ['timestamp', 'latitude', 'longitude', 'altitude']
trajectory3 = trajectory3.drop_duplicates(subset=['latitude', 'longitude'], keep='first')
trajectory3.reset_index(inplace=True)
trajectory3.drop('index', axis=1, inplace=True)
trajectory3.astype(float)
trajectory3['sum_of_coords'] = trajectory3['latitude'] + trajectory3['longitude']

trajectory4 = np.load( '/home/petar/Fakultet/Semester 8/Biologically inspired computing/Proekt/SampleTraj/Trj_3.npy' )
trajectory4 = pd.DataFrame(trajectory4)
trajectory4.columns = ['timestamp', 'latitude', 'longitude', 'altitude']
trajectory4 = trajectory4.drop_duplicates(subset=['latitude', 'longitude'], keep='first')
trajectory4.reset_index(inplace=True)
trajectory4.drop('index', axis=1, inplace=True)
trajectory4.astype(float)
trajectory4['sum_of_coords'] = trajectory4['latitude'] + trajectory4['longitude']
# targetPoint = -88
```

```python
# less than period
amp_window_size=24
# (maybe) as same as period
series_window_size=24
# a number enough larger than period
score_window_size=100
```

```python
spec = anom.Silency(amp_window_size, series_window_size, score_window_size)
```

# Trajectory 1

```python
score = spec.generate_anomaly_score(trajectory1['sum_of_coords'].to_numpy())
```

```python
fig = go.Figure()

fig.add_scatter(x=list(range(len(score))), y=score, name='biggest anomalies')

fig.show()
```

```python
# Plot data with estimated anomalous points
# plt.plot(trajectory1['sum_of_coords'].to_numpy(), alpha=0.5, label="observation")
index_changes = np.where(score > np.percentile(score, 99))[0]
# plt.scatter(index_changes, trajectory1['sum_of_coords'].to_numpy()[index_changes], c='green', label="change point")

fig = go.Figure()

fig.add_scatter(x=list(range(len(trajectory1['sum_of_coords']))), y=trajectory1['sum_of_coords'])

anomalies = trajectory1.iloc[trajectory1.index.isin(index_changes)]

fig.add_traces(go.Scatter(
    x=anomalies['sum_of_coords'].index, y=anomalies['sum_of_coords'], mode="markers", name="Ground Truth", hoverinfo="y"
))

fig.show()
```

```python
tmp = trajectory1.to_numpy()[:, 1:3]

avg_lat = np.mean([row[0] for row in tmp])
avg_lon = np.mean([row[1] for row in tmp])
```

```python
targetPoints = [260]
```

```python
m = folium.Map(
    location=[avg_lat, avg_lon],
    tiles="cartodbpositron",
    zoom_start=5
)

aline1 = folium.PolyLine(tmp, color='blue')
m.add_child(aline1)

# aline2 = folium.PolyLine(cTraj1, color='red')
# m.add_child(aline2)

for targetPoint in targetPoints:
    folium.Marker(location=[tmp[targetPoint][0], tmp[targetPoint][1]], icon=folium.Icon(color='red', icon_color='white', icon='male', angle=0, prefix='fa')).add_to(m)

# folium.Marker(location=[tmp[260][0], tmp[260][1]], icon=folium.Icon(color='red', icon_color='white', icon='male', angle=0, prefix='fa')).add_to(m)

for index, generated_point in anomalies.iloc[:, [1, 2]].iterrows():
#     print(generated_point)
    folium.Marker(location=[generated_point[0], generated_point[1]], icon=folium.Icon(color='darkblue', icon_color='white', icon='male', angle=0, prefix='fa')).add_to(m)


m
```
